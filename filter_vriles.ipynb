```python
# filter_vriles
#
# Removes vRILES on the last day and first two days of each month due to the erroneous NSIDC data shift (See Meier and Stewart 2019)
#
# Steven Cavallo
# December 2024
#
###########################
# imports
###########################
import matplotlib.pyplot as plt
import numpy as np
import sys, imp
import pandas as pd
import datetime as dt
from datetime import date
import os, datetime
import csv
from pathlib import Path
import gc


import weather_modules as wm
import utilities_modules as um

from mstats import *
```

```python
directory2filter = '/Users/scavallo/Documents/data/seaice_loss/paper_data_directory/'
fname2filter_prefix = 'rapid_seaice_loss_events*'
```

```python
files = Path(directory2filter).glob(fname2filter_prefix)
for file in files:
    gc.collect()
    
    filestr = str(file)
    fpath_cases_vrile_filtered = filestr[0:-4] + '_noshift.dat'

    fpath_cases_vrile = filestr
    aa = np.loadtxt(fpath_cases_vrile, skiprows=0)       
    datelist_vrile = aa[:,0]
    dextent_in_vrile = aa[:,1]
    dextent_in_nofilt_vrile = aa[:,2]    
    
    years_new_vrile = []
    months_new_vrile = []
    days_new_vrile = []
    mmdd_new_vrile = []
    yyyymmdd_new_vrile = []
    yyyymmddhh_new_vrile = []
    dextent_new_vrile = []
    dextent_new_nofilt_vrile = []

    ncases_vrile = np.size(datelist_vrile)
    tt = 0
    while tt < ncases_vrile:

        datestrinit_vrile = str(datelist_vrile[tt])	
        yyyy = datestrinit_vrile[0:4]
        mm = datestrinit_vrile[4:6]
        dd = datestrinit_vrile[6:8]
        hh = datestrinit_vrile[8:10]
        mmdd = datestrinit_vrile[4:8]
        yyyymmdd = datestrinit_vrile[0:8]
        yyyymmddhh = datestrinit_vrile[0:10]
 
        a = pd.Timestamp(int(yyyy), int(mm), int(dd))
        month_end_check = a.is_month_end
    
        day_start_check = int(dd)
        if ( (month_end_check == False) and (day_start_check != 1) and (day_start_check != 2) ):
    
            years_new_vrile.append(yyyy)
            months_new_vrile.append(mm)
            days_new_vrile.append(dd)
            mmdd_new_vrile.append(mmdd)
            yyyymmdd_new_vrile.append(yyyymmdd)
            yyyymmddhh_new_vrile.append(yyyymmddhh)
            dextent_new_vrile.append(dextent_in_vrile[tt])
            dextent_new_nofilt_vrile.append(dextent_in_nofilt_vrile[tt]) 
        
        tt+=1

    outfile = open(fpath_cases_vrile_filtered,'w')
    wcount = 0
    for ii in range(0,len(yyyymmddhh_new_vrile)):
        print(yyyymmddhh_new_vrile[ii],dextent_new_vrile[ii],dextent_new_nofilt_vrile[ii])
        datesprintnow = np.array(yyyymmddhh_new_vrile[ii],dtype='int')
        outfile.write('%-10s %7.4f %7.4f\n' % (datesprintnow, dextent_new_vrile[ii], dextent_new_nofilt_vrile[ii] ))
```
